```python
from data_loader import load_data

path = '../Data/processed/preprocessed_data.csv'
df = load_data(path)
df.head()
```

```python
df ["tags"] = df['genres'] + ' ' + df['keywords'] + ' ' + df['production_companies'] + ' ' + df['crew'] + ' ' + df['tagline'] + ' ' + df['overview']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df ['popularity_scaled'] = scaler.fit_transform(df[['popularity']])
df ['vote_average_scaled'] = scaler.fit_transform(df[['vote_average']])
df ['budget_scaled'] = scaler.fit_transform(df[['budget']])
df ['revenue_scaled'] = scaler.fit_transform(df[['revenue']])
```

```python
df.to_csv('../Data/Ready/final_data.csv', index=False)
```
