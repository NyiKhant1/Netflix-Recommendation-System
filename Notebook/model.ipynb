```python
import sys
sys.path.append('../SRC')
from data_loader import load_data   
```

```python
path = '../Data/Ready/final_data.csv'

df = load_data(path)

print (df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x = df['tags'].fillna('').values
x = vectorizer.fit_transform(x)
similarity = cosine_similarity(x)
print (similarity)  
```

```python
import pickle

pickle.dump(df, open('../Data/Model/movies.pkl', 'wb'))
pickle.dump(similarity, open('../Data/Model/similarity.pkl', 'wb'))
```
